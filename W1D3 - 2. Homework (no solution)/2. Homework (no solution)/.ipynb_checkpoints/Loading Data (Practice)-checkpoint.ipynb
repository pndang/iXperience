## <font color='#eb3483'> 3. Copy paste </font>

Once again, our old favorite pandas is coming to the resuce! Pandas has some excellent functionality for getting data from websites into dataframes.    
To get data from our clipboard (i.e. after you copy something from a webpage). Pandas has a function that can read what we have copied to the clipboard and turn it into a dataframe.

For example, we can go to [this page](https://www.worldatlas.com/articles/largest-cities-in-europe-by-population.html) that has a table containing a list of population by city in Europe. 

Copy the table using plain old highlight and copy (windows = ctrl+c, mac = cmd+c, or right click and copy).

We can copy the table and read it as a dataframe with `pd.read_clipboard`.

```python
df = pd.read_clipboard()
```
