So what's our first step? 

## <font color='#eb3483'> 1. Test statistic </font>
We need to identify the metric we want to test.

The metric we want to compare is whether people converted or not. So our dependent variable will be converted.

The independant variable will be the group (whether or not they were in the control (old website) or treatment (new website) groups.

```python
#Let's pull out the conversion stats for each group

df[df.group == "control"].describe()
```

```python
df[df.group == "treatment"].describe()
```

Look fairly similar so far. Let's make a plot to have a look at the difference in conversion rate.

```python
sns.barplot(x=df.group, y=df.converted)

```

Hmm look fairly similar - but eyeballing doesnt cut the mustard. So let's formulate our Null Hypothesis

## <font color='#eb3483'> 2. Test hypotheses </font>


We don't have a guess which website will perform better so we'll choose a two-tailed test. 
where $p$ and $p_0$ stand for the conversion rate of the new and old design, respectively.

We are trying to show that the websites being the same isnt the case. Therefore our null hypothesis is:

Null hypothesis ($H_0): p = p_0$  

Alternate hypothesis ($H_a): p \ne p_0$

## <font color='#eb3483'> 3. Compute the probability </font>


We'll set a **confidence level of 95%**:

$$\alpha = 0.05$$

This means  that whatever conversion rate we observe for our new design in our test, we want to be 95% confident it is statistically different from the conversion rate of our old design, before we decide to reject the Null hypothesis $H_0$.

The $\alpha$ value is a threshold we set, by which we say "if the probability of observing a result as extreme or more ($p$-value) is lower than $\alpha$, then we reject the null hypothesis". 

<div>

</div>

Since we have two samples and we have a sample size greater than 30, we can use the normal approximation for calculating our 𝑝 -value (i.e. z-test).

Python makes all the calculations very easy. 

We can use the <code>ztest</code>
from <code>statsmodels.stats.weightstats</code>. 


```python
!{sys.executable} -m pip install statsmodels
from statsmodels.stats.proportion import proportions_ztest, proportion_confint
```

```python
# break our data set into our two samples 
# and pull out the converted results
control_group = df[df.group == "control"]["converted"]
control_group.head()
```

```python
treatment_group = df[df.group == "treatment"]["converted"]
treatment_group.head()
```

```python
from statsmodels.stats.weightstats import ztest as ztest
```

```python
z_stat, pval = ztest(x1=control_group, x2=treatment_group, value=0)
```

```python
z_stat
```

```python
pval
```

<div>

</div>

Since our $p$-value=0.216 is way greater than our $\alpha$=0.05, we cannot reject the null hypothesis $H_0$ which said the websites are the same, which means that the conversion rate on our new website was not significantly different than the old.

Don't forget ...

When doing a significance test, there two possible types of error:

__Type I error__
> You reject the null hypothesis in favor of the alternative hypothesis, but the null hypothesis was actually true. Also know as a false positive. With a $\alpha$ = 0.05 significance level, you expect this type of error to happen 5% of the time. The higher the $\alpha$, the less likely you are to make this error.


__Type II error__
>You do not reject the null hypothesis in favor of the alternative hypothesis, but the alternative hypothesis was actually true. Also know as a false negative. The lower the $\alpha$, the less likely you are to make this error.
