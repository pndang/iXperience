### <font color='#eb3483'> Exercise 1 </font>
Calculate summary statistics (mode, mean, median, IQR, skewness, kurtosis, standard deviation) for the variable age. What are its quartiles for women?

```python
survey.age.describe()
```

```python
stats.mode(survey.age)
```

```python
survey.age.skew()
```

```python
survey.age.std()
```

```python
for qt in [0.25, 0.5, 0.75]:
    print("Quantile {}: {}".format(qt, 
            survey.age[(survey.gender=="Female")&(survey.age.notnull())].quantile(qt)
                                 ))
```

*Note*: You could also find the quartiles by doing:

```python
survey[survey.gender=="Female"].age.describe()
```

### <font color='#eb3483'> Exercise 2 </font>
Which region in Portugal has the highest average internet usage?

```python
survey.groupby("region")["internet_usage_minutes"].mean()
```

We see that Alentejo has the highest average internet usage

### <font color='#eb3483'> Exercise 3 </font>

Which party did the oldest person/people vote for? How about the youngest one/s? 

```python
survey[survey.age==survey.age.max()].party_voted_last_election
```

```python
survey[survey.age==survey.age.min()].party_voted_last_election
```

Oh, we see, the youngest people in the dataset werent of voting age in 2015!. Given that this dataset was gathered in 2016, we can assume the youngest respondents capable of voting must be older than 18 years old, so at least 19 years old.

```python
survey[survey.age==19].party_voted_last_election.value_counts()
```

So we see that most young people (Millenials!!) didnt vote, except one who voted for *PS - Partido Socialista*

### <font color='#eb3483'> Exercise 4 </font>
What is the correlation between age and Internet usage?

```python
survey[['age','internet_usage_minutes']].corr()
```

```python
survey.corr()["age"]["internet_usage_minutes"]
```

*Note*, using `np.corrcoef` only works if there arent NaN values

```python
np.corrcoef(survey.age, survey.internet_usage_minutes)
```

### <font color='#eb3483'> Exercise 5 </font>
Which variables are highly correlated? Bonus: can you plot your results in a heatmap?

```python
survey.select_dtypes([int, float]).corr()
```

```python
correlations = survey.select_dtypes([int, float]).corr()
correlations[(correlations>0.3) & (correlations<1)].dropna(how="all", axis=1)
```

```python
sns.heatmap(survey.select_dtypes([int, float]).corr())
```
